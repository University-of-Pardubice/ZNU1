# Module, package, library

## Module
A module is a .py file with functions, classes, and variables.

It allows separating the program logic into independent parts.

A module is referenced with the import statement

````python
# Example: The file my_module.py would contain:
# def greetings(name):
#     print(f"Hello {name}")
````

````python
# my_module.py must be located in the same directory:
# import my_module
# my_module.greetings("Eva")
````

## Package
* A folder that contains multiple modules, packages, etc.
* The directory structure on disk is the structure of the package
* \_\_init\_\_.py - a predefined name for the initialization module
* Modules can have classes with the same name; in that case you need to import or qualify them correctly.
* You can import a specific module from a package.

If a module is missing, it can be installed using pip:

**pip install numpy pandas matplotlib**

Note: `collections`, `os`, `re`, `sys` and others are **built-in modules** of Python — they are part of the installation and do not need to be installed.

## Exploring the contents of a module

The `dir()` function returns a list of all attributes and functions of a module. `help()` shows the documentation.

````python
import os

print(dir(os))          # List of all module attributes and functions
print(os.__file__)      # path to the module file on disk
````

````python
help(os.path.join)      # documentation of a specific function
````

## The __name__ pattern

* Every module has a `__name__` attribute. If the file is **run directly**, its value is `"__main__"`. 
* If it is **imported** by another module, the value is the module's name.
* In a Jupyter notebook, __name__ is always "__main__"


The pattern **if __name__ == "__main__"** is often used. This pattern is used so that test or startup code in a module does not run when it is imported.

````python
print(__name__)

# Typical usage in a .py file:
#
# def main():
#     print("Run directly")
#
# if __name__ == "__main__":
#     main()          # runs only when the file is run directly
#                     # does not run when imported
````
