# Code blocks and indentation
* Code blocks are used to group multiple statements, for example in conditions, loops, functions, or class definitions.
* Unlike other languages, which use {}, begin...end, etc., **Python uses indentation**.
* Incorrect indentation is one of the most common beginner mistakes in Python.

## Indentation rules
* You create a new block by indenting the lines under a condition, loop, or function.
* The same number of spaces or tabs must be used for all lines in a block.
* Mixing tabs and spaces, or incorrect indentation, will cause an error.
* **By default, 4 spaces are used for one indentation level.**

````python
a=10
if (a == 1):
    # begin block
    print ("A is 1")
    print ("Next line of code")
    # end block
elif (a==2):
    # begin block
    print ("A is 2")
    print ("Next line code")
    # end block
else:
    # begin block
    print ("A is not 1 nor 2")
    # end block
````

# Conditions
- Conditions are used to decide which statements should be executed.
- Conditions can use the classic logical operators **and**, **or**, **not**.
- Conditions can also be compound. In that case they are separated with parentheses.

````python
x = 10
y = 5

if x > 0 and (y < 10 or y == 5):
    print("The condition is fulfilled.")
````

The **in** operator is used to check whether an element is in a list, etc.

````python
fruits = ['apple', 'banana', 'cherry']
if 'banana' in fruits:
    print("banana is in list")
````

The **is** operator compares whether two variables refer to **the same object** in memory (not just the same value). It is most commonly used to compare with `None`.

````python
x = None
if x is None:
    print("x is None")        # correct way to compare with None

a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)   # True  – b points to the same object
print(a is c)   # False – c has the same value, but is a different object
print(a == c)   # True  – the values are equal
````

The **ternary operator** (conditional expression) lets you write a simple condition on a single line:
```
value_if_true  if  condition  else  value_if_false
```

````python
age = 20
status = "adult" if age >= 18 else "kid"
print(status)

# equivalent using if/else:
# if age >= 18:
#     status = "adult"
# else:
#     status = "kid"
````

The **match** statement is the equivalent of the switch statement in other languages. Available since version 3.10

````python
number = 1
match (number):
    case 1:
        print ("Number 1")
    case 2:
        print ("Number 2")
    case _:
        print ("A number other than 1 or 2")
````

# For loop
The **for** loop is used to repeatedly execute statements for each element of some sequence (list, string, range, etc.).
* There is a colon after the for definition.

**range** generates a sequence of numbers, usually for iteration:
* if not specified, it starts at 0
* it ends at a number one less than the input parameter
* range(stop)         # 0, 1, 2, ..., stop-1
* range(start, stop)  # start, start+1, ..., stop-1
* range(start, stop, step)  # step of size step

````python
for i in range(4):
    print (i)
````

Even numbers

````python
for i in range(2, 10, 2):
    print(i)
````

Countdown

````python
for i in range(5, 0, -1):
    print(i)
````

````python
for i in range (1, 5):
    if (i % 2 != 0):
        print (f"Number {i} is odd")
    else:
        print (f"Number {i} is even")
````

# while loop
The **while** loop repeats a block of statements as long as the condition is true (True).

````python
i=1
j=5
while i < j:
    print(i)
    i += 1
````

# The break, continue, and else statements
The **break** statement immediately terminates the entire loop, regardless of whether there are remaining iterations. It is typically used when searching for something or terminating an infinite loop.

````python
for i in range(5):
    if i == 3:
        break
    print(i)
````

The **continue** statement ends the current iteration. The rest of the loop body is not executed, but the loop continues with the next iteration.

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````

````python
for i in range(1000):
    if i % 3 == 0:
        print(f'{i} is divisible by 3')
        continue

    if i > 10:
        print(f'{i} > 10, therefore the end')
        break
        
    print(i)
````

The **pass** statement means do nothing. It is often used in prepared branches of conditions, functions, ..., that we will implement later, but we want the program to be valid.

The **break** statement is often used when what was being searched for has been found during the loop.

The **else** statement is then used for cases when nothing was found, i.e. when the loop was not terminated with break.


````python
numbers = [1, 2, 3 ,4, 5]
for i in numbers:
    if i == 100:
        print('I found 100')
        break
else:
     print('100 not found')
````

# Assert
* **assert** is used to check whether a condition is true during program execution.
* If the condition is not met, Python raises **AssertionError**.
* Typically used to check assumptions in code (debugging, tests).
* assert is often used in unit tests or to check function input parameters.
* If you run Python with the -O option (optimization), assert statements are ignored.

If the condition returns True, then nothing happens.

````python
x = "hello"
assert x == "hello"
````

If the condition returns False, an AssertionError is raised:

````python
assert x == "goodbye"   # ← this cell will end with an AssertionError
````
