# Data Types, Classes

* Python is a **strongly typed** language. For example, you cannot add incompatible types without conversion (e.g. number + string).
* Python is also **dynamically typed**. The type of a variable is determined by its value at the first assignment; type checking happens only at runtime, and explicit casting is possible.

```
    x = "123"
    y = int(x) 
```

* Everything is an object – even numbers, functions, types, ... are objects of some class.

The basic built-in types are as follows:
* Numeric types:
   * int – integers, not limited to a fixed number of bits, you can work with very large numbers: x = 451654872154787894564156475645564654
   * float – decimal numbers, support special values: inf, -inf, NaN (Not a Number)
   * complex – complex numbers, e.g. 2 + 3i
* Strings:
   * str – strings are stored in unicode. Beware, if we work with text containing national characters, it sometimes may not have the correct code page.
   * bytes - binary data, e.g. b'\x68\x65\x6c\x6c\x6f' (represents the text "hello")
* Boolean values:
   * bool – values True and False
   * None - an empty value, in other languages e.g. null
* Sequences and collections:
   * list – list,
   * tuple – tuple (immutable),
   * set – set,
   * dict – dictionary (key–value).

Here are some examples of entering values.

````python
int1 = 1
int2 = int1 + 1
int3 = 1_000_000
float1 = 1.0
float2 = float1 + 1
string1='0'
````

Because a variable's type is determined dynamically at the first assignment, it can sometimes be useful to find out what type a variable actually is. The **type** command is used for this.

````python
print (type(int1))
print (type(int2))
print (type(float1))
print (type(float2))
print (type(string1))
````

You can also use the **isinstance** function to test whether a variable is of a certain type.

````python
print (isinstance(int2, int))
print (isinstance(float1, str))
````

The type of a variable is determined from the given value according to built-in rules.

Sometimes, to avoid errors in the program, it can be useful to define the data type explicitly.

Or we may want to cast the input value.

````python
int3 = int (1.0)
float3 = float(1)
string2 = str (int3)

print (type(int3))
print (type(float3))
print (type(string2))
````

**For immutable data types, a copy of the data is made on assignment.** A new instance of the data is created in memory, and each variable points to a different part of memory.
* bool
* int
* float
* complex
* str
* bytes

````python
x = 1
y = x
x = 2
print ("x:", x)
print ("y:", y)
````

## Integers
We will very often work with integers, so let's look at the basic operations.

When converting to an integer, Python truncates, it does not round.

````python
print (int(3.9))
print (int(-2.9))
````

If we want to round, we must use the **round()** function. The round() function uses banker's rounding – to the nearest even number.

````python
print (round(6.5))        # 6
print (round(7.5))        # 8
print (round(3.14, 1))    # 3.1  (to 1 decimal place)
print (round(2737, -2))   # 2700 (to hundreds)
````

Standard operators +, -, *, / are used for working with integers.

We add these to them:
* // – integer division. For positive numbers it "truncates" the decimal part; for negative numbers it returns the lower value (e.g. -7 // 2 == -4); if either operand is a float, the result is also a float.
* ** – exponentiation
* % – modulo (remainder after division)

````python
print(7 // 2)   # 3
print(-7 // 2)  # -4
print(2 ** 3)   # 8
print(7 % 3)    # 1
````

Sometimes we will want to convert between number systems.

````python
print(int("FF", base=16))     # 255
print(int("0b10010", base=2)) # 18

print(bin(99))   # '0b1100011'
print(hex(99))   # '0x63'
print(oct(99))   # '0o143'
````

## Strings

* Python uses Unicode (UTF-8) for text.
* One character can be stored in 1–4 bytes.
* The program's source file (program.py) must be saved in UTF-8.
* Strings can be written using quotes:
   * double "..."
   * or single '...'

A string can be split into variables.

````python
s = 'cool'
a, b, c, d = 'cool'
print (d)
````

Basic operations with strings include getting the length of a string, accessing the n-th character, concatenating strings, and repeating characters.

````python
string_length = len(s) 
first_char = s[0]
string_join = 'abc' + 'def'
multiple_a = 'a' * 4
print (string_length)
print (first_char)
print (string_join)
print (multiple_a)
````

Sometimes we will want a string to contain multiple lines. The special character **\n** is used as a separator.

````python
multiline='first\nsecond\nthird\n'
print (multiline)
````

If we want to use a sequence of characters like \n, \t, ... in a string so that they are not interpreted, we have three options:
* **raw string** – prefix `r` before the quotes (the most common way, suitable for paths and regular expressions)
* the **repr** function – returns the representation of the string including quotes
* doubling the backslash **\\\\**

````python
print (r"string\n")       # raw string: string\n
print (repr("string\n"))  # repr:        'string\n'
print ("string\\n")       # doubling:    string\n
````

### String formatting
When printing variables, we will want to control their appearance.

The older way of formatting uses the **.format()** method, whose first input is the format string, followed by the variables to be printed.

* At the place where a variable should be printed, its order number (numbered from 0) is given in {}.
* If the order is clear, it can be omitted.
* Variables can also be named.

````python
a = [1, 2, 3]
b = 'xxx'
c = 1.222342

print("b {0} a {1[0]}".format(b, a))
print("b {} a[0] {}".format(b, a[0]))
print("b {b} a {a}".format(b=b, a=a[0]))
````

For a nicer printout, you can specify the alignment and length of the string.

````python
print ('{:<30}'.format('left aligned'))
print ('{:>30}'.format('right aligned'))
print ('{:^30}'.format('centered'))
print ('{:*^30}'.format('centered'))
````

When printing numbers, it is good to define their format.
* Data type (**f**loat, **d**ecimal)
* Number of digits before and after the decimal point
* Thousands separator

````python
print ('{0:.1f}'.format(c))
print ('{0:03d}'.format(10))
print ('{0:,}'.format(123456789))
````

For clearer formatting, a newer, clearer method is used: **f-string**. There is an f before the string, and inside the string we can use variables, but also functions, methods, etc.


* **f'...'** – inside we can directly use variables {b}, {c}
* **str.format()** – suitable for more complex formatting or repeated use

````python
print (f'The value of b is {b}, the value of c is {c}')
````

````python
text = 'Python'
print(f'{text:>10}')  # right: '    Python'
print(f'{text:^10}')  # center: '  Python  '
print(f'{text:*^10}') # center with asterisks: '**Python**'
````

````python
x = 42
X = 1254684
print(f'{x:05d}')   # zero-padded: 00042
print(f'{X:,}')     # thousands separated by comma: 1,254,684
````

````python
y = 12345.6789
print(f'{y:.2e}')   # 1.23e+04
````

````python
a = 5
b = 3
print(f'{a} + {b} = {a+b}')  # 5 + 3 = 8
````

When printing multiple variables, you can change the separator instead of a space. This can be useful, for example, when creating text files.

````python
print ("A", "B", "C", sep="|")
````

You can also redefine the character printed at the end.

````python
print ("A", end=",")              # a , is printed at the end of the statement
print ("B")                       # by default, \n (a new line) is printed at the end of the statement
````

# Mutable data types
Unlike immutable types (int, str, tuple …), mutable data types can be modified after creation.

- list - an ordered collection of elements, which can be of different types. Elements can be added, removed, or changed.
- set - an unordered collection of unique elements. Suitable for working with set operations (union, intersection).
- dict - a collection of key : value pairs. Keys must be immutable (e.g. numbers, strings, tuples). Values can also be mutable.

**For mutable data types, assigning a value does not make a copy of the data as it does for immutable types.** A pointer to the data structure is passed instead.

````python
a=[1,2,3]
b=a        # creates a copy of the reference, pointer
b[2]=4     # a[2] will also be 4

print (a)
print (b)
````

## List
* A list of arbitrary variables, but it is advisable to use the same data type or values of the same meaning.
* Elements can be changed, added, and removed over time.

**For classic multidimensional arrays we will use the numpy library.** We will cover numpy later.

### Creating a list

````python
value = 6
a_list = ['c', 'd', 1, True, value]
print (a_list)
````

### Accessing elements
* Indexing starts at 0.
* Indexing with negative numbers starts from the end.

````python
print (a_list[0])     # first element
print (a_list[-1])    # last element
print (a_list[0:3])   # first 3 elements
print (a_list[0:6:2]) # every second element from index 0 to 5
print (a_list[:])     # copy of the list
print (a_list[::-1])  # reverse the list

a_list[2:4]='ab'      # assign a value to a slice of the list
print (a_list)
````

### Adding elements

````python
a_list.append(8)           # adds one element at the end
a_list.append([3,4])       # adds a list as a single element
a_list.extend([3,4])       # adds all the elements of the list
a_list.insert(0, 6)        # inserts an element at index 0

print (a_list)
````

### Concatenating lists

````python
a_list = a_list + [1, 2.0]
print (a_list)
````

### Removing elements

````python
del a_list[1]                   # by index
print (a_list)
a_list.remove(6)                # by value (first occurrence)
print (a_list)
print (a_list.pop())            # removes and returns the last element
print (a_list.pop(1))           # removes the element at index 1 and returns it
print (a_list)
````

### List information

````python
print (len(a_list))             # length of the list
print (a_list.count(1))         # number of occurrences of the value
print (6 in a_list)             # True/False
print (a_list.index('d'))       # index of the element, raises an exception if not found
````

### Further operations, sorting and splitting

````python
num_list = [56, 8, 0, -5]
num_list.sort()                     # sort the list (elements must be of the same type)
print (f'Sorted list {num_list}')

head, *middle, tail = [1,2,3,4,5]   # split the list
print (f'Head {head}')
print (f'Middle {middle}')
print (f'Tail {tail}')
````

### Comparing lists
The comparison depends on order. If order should not matter for the comparison, we must sort the list before comparing.

````python
l1 = [10, 20, 30, 40, 50]
l2 = [20, 30, 50, 40, 70]
l3 = [50, 10, 30, 20, 40]

if l1 == l2:
    print ("The lists l1 and l2 are the same.")
else:
    print ("The lists l1 and l2 are not the same.")

if l1 == l3:
    print ("The lists l1 and l3 are the same.")
else:
    print ("The lists l1 and l3 are not the same.")
````

````python
l1 = [10, 20, 30, 40, 50]
l2 = [20, 30, 50, 40, 70]
l3 = [50, 10, 30, 20, 40]

l1_sorted = sorted(l1)
l2_sorted = sorted(l2)
l3_sorted = sorted(l3)

if l1_sorted == l2_sorted:
    print ("The lists l1 and l2 are the same.")
else:
    print ("The lists l1 and l2 are not the same.")

if l1_sorted == l3_sorted:
    print ("The lists l1 and l3 are the same.")
else:
    print ("The lists l1 and l3 are not the same.")
````

### Copying a list
Because list is a mutable data type, assigning it to a new variable does not create a copy, but only creates a new pointer.

````python
a=[1,2,3]
b=a
b[2]=4

print (a)
print (b)
````

Sometimes it can be useful to create a copy of a data structure. The first way of copying may look a bit odd.

````python
c=a[:]
print (a)
print (c)
print ("-----------")
c[0]=0
print (a)
print (c)
````

We can use the **copy** function to create a copy of a list.

````python
a=[1, 2]
b=a.copy()
b[0]=3
print (a)
print (b)
````

Beware, copy only makes a copy down to the first level of the list.

````python
a=[[1,2], 3]
b=a.copy()
b[0][0]=5
print (a)
print (b)
````

If the structure is more complex, we must use **deepcopy**

````python
import copy
a=[[1,2], 3]
b=copy.deepcopy(a)
b[0][0]=5
print (a)
print (b)
````

## Set

* An unordered collection of unique values.
* Each element can occur only once.
* Suitable for operations such as intersection, union, difference.

### Creation

````python
s = {1, 2, 3, 4}
s2 = set([3, 4, 5, 6])   # convert from a list
empty_set = set() 
print (s) 
print (s2)
````

### Adding and removing elements

````python
s.add(5)                # adds an element
s.update ({3, 4, 5})    # adds multiple values to the set, can be given as a set or a list
s.update ([3, 4, 5])    # adds multiple values to the set, can be given as a set or a list, it doesn't matter
s.remove(2)             # removes an element, raises an error if it doesn't exist
s.discard(10)           # removes an element if it exists, otherwise does nothing
s.pop()                 # removes a random value and returns it
print (s)
````

### Operations with sets

````python
print (f"Union {s.union(s2)}")                                  # union
print (f"Intersection {s.intersection(s2)}")                    # intersection
print (f"Difference {s.difference(s2)}")                        # difference (elements in s that are not in s2)
print (f"Symetric difference {s.symmetric_difference(s2)}")     # elements that are in one set or the other, but not both
print (f"Subset {s.issubset(s2)}")                              # subset test
print (f"Superset {s.issuperset(s2)}")                          # superset test
print (f"Length of set s {len(s)}")                                 # length of the set
````

# Binding objects and names
* An object can have one or more names (variables) that refer to it.
* A name does not denote a memory location – it is just a reference to an object.
* Any object can be assigned to a variable, and the type of the assigned object can be changed later.

````python
x=10
x='python'
print (x)
````

* Python does not support explicit type declarations.
* Since version 3.5, **type hints** can optionally be used, which help with static analysis and code readability.
* Type hints do not prevent assigning a different type; they only serve for checking (e.g. in an IDE or linter).

````python
def greetings(name: str) -> str:
    return f"Hello, {name}!"

print(greetings("world"))   # works correctly
print(greetings(42))        # the type hint is not enforced at runtime, Python does not raise an error
````

# Deleting a variable
Deleting a variable removes the binding between the name and the object, not the object itself.

````python
x = 10
del x   # x no longer exists
````

If an object has no references (no names point to it), the garbage collector automatically removes it and frees the memory.

Python manages memory automatically, so there is no need to explicitly free objects.

````python
y = [1, 2, 3]
z = y        # now there are two references to the list
del y        # the reference z remains, the object is not deleted
del z        # no references → the garbage collector frees the memory
````
