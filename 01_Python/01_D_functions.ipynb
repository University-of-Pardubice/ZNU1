* Assign the output to a variable.
* Use a function as a parameter of another function.
* Create it at runtime.

````python
def foo(bar):
    """ Documentation string
        on multiple lines
    """
    
    return bar
````

````python
f = foo(1)
print (f)
print (foo("B"))
````

# Arguments

Functions in Python can have **default parameter values**. If the user does not provide this parameter when calling the function, this default value is used.

**Watch out for mutable objects!**

Do not use lists, dictionaries, sets, etc. as a default value, because they are created only once when the function is defined and are then shared across all calls.
This can lead to bugs.

````python
def bar(a=1, b=2):
    """# default parameter values"""
    return a + b
bar()
````

Default parameters can be omitted or overridden.

Named parameters can be reordered.

````python
print (bar(b=3))
print (bar (b=3, a=5))
````

***args** allows a function to accept any number of positional arguments.

Inside the function, args behaves like a tuple.

````python
def sum (*args):
    """ Using any number of parameters """
    result = 0
    for arg in args:
        result += arg
    return result
sum (1, 2, 3)
````

****kwargs** works similarly to ***args**, but instead of a tuple it stores the arguments in a dictionary.

It is used for any number of named (keyword) arguments.

````python
def info(**kwargs):
    print(kwargs)

info(name="Eva", age=25, city="Brno")
````

You can use concurrent arguments, *args and **kwargs.

The order should be:
* regular parameters
* *args
* **kwargs

````python
def universal(required, *args, **kwargs):
    print("Required parameter:", required)
    print("Other (args):", args)
    print("Key (kwargs):", kwargs)    

universal("start", 1, 2, 3, name="Eva", age=25)
````

````python
person = {"name": "Karel", "age": 40}
universal("start", **person)
````

# Return value

A function can return multiple values. They are returned as a tuple.

A tuple can also be split into individual variables.

````python
def division(a, b):
    fraction = a // b
    remainder = a % b
    return fraction, remainder    

result = division (17, 5)
print(result)    
print(type(result))  

a, b = division (10, 3)
print (a)
print (b)
````

# Lambda functions

A **lambda** is an anonymous function written on a single line. It is suitable for simple operations passed as an argument to another function.

```
lambda parameters: expression
```

* It can have any number of parameters, but only one expression.
* The result of the expression is returned automatically (no `return` needed).
* For more complex logic, use a named function instead.

* Syntax: lambda parameters: expression
    * parameters → the function's inputs (0 or more)
    * expression → the value the function returns

````python
square = lambda x: x ** 2
print(square(5))    # 25

add = lambda a, b: a + b
print(add(3, 4))    # 7
````

````python
numbers = [1, 2, 3, 4]
twice = list(map(lambda x: x*2, numbers))
print(twice)
````

Lambda functions are most commonly used as an argument to `sorted()`, `map()`, and `filter()`.

````python
people = [{"name": "Eva", "age": 25}, {"name": "Jan", "age": 19}, {"name": "Petr", "age": 32}]

# sorted – sorted by age
sorted_people = sorted(people, key=lambda p: p["age"])
print(sorted_people)

# map – applies a function to each element and returns an iterator
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)    # [1, 4, 9, 16, 25]

# filter – keeps only the elements that meet the condition
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)      # [2, 4]
````

# Data type hints
Python has the option of type hints. These are used for readability and for tools (linting, IDE, mypy), but Python does not check them at runtime.

After the colon : we state the expected type of the parameter, after -> the return value type

````python
def greet(name: str, age: int) -> str:
    return f"Hello, {name}, you are {age} years old."

print(greet("Petr", 25))
````

The function runs and returns the result. Some IDEs will report an error or warning, because the age is given as a string.

````python
print(greet("Petr", "25"))
````

They are used from the typing module (since Python 3.9 also directly list, dict, etc.):

````python
from typing import List, Dict, Tuple, Optional

def sum_list(values: List[int]) -> int:
    return sum(values)

def get_user() -> Dict[str, str]:
    return {"name": "Anna", "role": "admin"}

def position() -> Tuple[int, int]:
    return (10, 20)

def maybe_number(flag: bool) -> Optional[int]:
    return 42 if flag else None
````

You can also use the more general types Any, Union, Literal
* Any = can be anything
* Union = multiple types
* Literal = specific allowed values

````python
from typing import Any, Union, Literal

def print_anything(x: Any) -> None:
    print(x)

def parse_num(s: str) -> Union[int, float]:
    return int(s) if s.isdigit() else float(s)

def set_mode(mode: Literal["r", "w", "a"]) -> None:
    print(f"I open the file in mode {mode}")
````

Hints for *args and **kwargs

*args and **kwargs can be typed similarly to other parameters – you just need to remember that:
* *args = a tuple of arguments (→ Tuple[type, ...])
* **kwargs = a dictionary (→ Dict[str, type])

````python
from typing import Tuple, Dict

def add_all(*args: int) -> int:
    return sum(args)

def print_info(**kwargs: str) -> None:
    for key, value in kwargs.items():
        print(f"{key}: {value}")
````

# Scope

* The evaluation priority follows the LEGB rule. In Python, when the interpreter encounters a variable, it looks for it in this order:
- L - Local
- E - Enclosing
- G - Global
- B - Built-in

````python
# Local
def f():
    x = 10  # Local
    print(x)
````

````python
# Enclosing
def outer():
    x = "enclosing"
    def inner():
        print(x)  # searches within the "outer" range
    inner()

outer()
````

````python
# Global
x = "global"
def f():
    print(x)  # takes a global variable

f()
````

````python
# Built-in
def f():
    print(len([1,2,3]))  # "len" is a built-in function

f()
````

# Explicit global, nonlocal declaration

**global** tells Python that we want to use a global variable (defined at module level). Otherwise, assignment would create a new local variable.

**nonlocal** is used inside a nested function. It says: use the variable from the enclosing function. Useful for working with closures.

````python
x = 5  # global

def f():
    global x      # Use of a global variable
    x = 10        # global x will change
    print("inner:", x)

f()
print("outer:", x)
````

````python
def outer():
    x = "origin"
    def inner():
        nonlocal x    # Using the variable z outer()
        x = "changed"
    inner()
    print(x)

outer()
````
