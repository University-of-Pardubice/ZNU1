# zip() — iterating over two lists at once

`zip()` combines two (or more) lists into pairs and lets you iterate over them together.

Ugly way

```python
names = ["Alice", "Bob", "Carol"]
scores = [95, 82, 78]

for i in range(len(names)):
    print(names[i], scores[i])
```

Nice way

```python
names = ["Alice", "Bob", "Carol"]
scores = [95, 82, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

# any() and all()

`any()` returns `True` if at least one element satisfies the condition.
`all()` returns `True` if all elements satisfy it.

Ugly way

```python
values = [0, 1, 2, 3]

found = False
for v in values:
    if v > 2:
        found = True
        break

print(found) 
```

Nice way

```python
values = [0, 1, 2, 3]

found = any(v > 2 for v in values)
all_positive = all(v >= 0 for v in values)

print(all_positive)
```

# dict.get() — safe access to a dictionary

Accessing via `d[key]` raises `KeyError` if the key doesn't exist. `d.get(key, default)` returns a default value instead.

Ugly way

```python
config = {"host": "localhost", "port": 8080}

# ugly – raises KeyError if key is missing
timeout = config["timeout"]
```

Nice way

```python
config = {"host": "localhost", "port": 8080}

timeout = config.get("timeout", 30)   # returns 30 if key doesn't exist
host    = config.get("host", "127.0.0.1")

print(timeout)  # 30
print(host)     # localhost
```

# enumerate() with start=

`enumerate()` adds a counter to the iteration. The `start=` parameter sets the initial value (default is 0).

Ugly way

```python
items = ["apple", "banana", "cherry"]

for i in range(len(items)):
    print(f"{i+1}. {items[i]}")
```

Nice way

```python
items = ["apple", "banana", "cherry"]

for i, item in enumerate(items, start=1):
    print(f"{i}. {item}")
```

# The walrus operator :=

The **walrus operator** (`:=`) assigns a value to a variable and simultaneously returns it as the result of the expression. A typical use is in a `while` loop — assigning and testing the condition in one step.

Ugly way

```python
data = b"hello world from a long stream"
i = 0
block_size = 8
blocks = []
while True:
    block = data[i:i+block_size]
    if not block:
        break
    blocks.append(block)
    i += block_size
print(blocks)
```

Nice way

```python

data = b"hello world from a long stream"

# Beautiful – walrus assigns a block and checks at the same time to see if it is empty
import io
f = io.BytesIO(data)
blocks = []
while block := f.read(8):
    blocks.append(block)
print(blocks)
```

# Caching function results (memoization)
Sometimes a function calls itself or performs expensive computations.

To avoid processing the same inputs multiple times, we can store the results in a cache.

```python
def fib(n: int) -> int:
    if n <= 1:
        return n
    else:
        return fib (n - 1) + fib (n - 2)
```

```python
for i in range (0, 40):
    print (f'{i}: {fib(i)}')
```

Python has a built-in @functools.lru_cache decorator:

LRU = Least Recently Used (removes the least recently used items once the cache reaches its limit)

It allows automatically caching results based on the function's parameters.

```python
from functools import lru_cache
@lru_cache
def fib(n: int) -> int:
    if n <= 1:
        return n
    else:
        return fib (n - 1) + fib (n - 2)
```

```python
for i in range (0, 40):
    print (f'{i}: {fib(i)}')
```
