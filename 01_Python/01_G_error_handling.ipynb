# Exceptions

Python allows you to catch and react to errors during program execution, so-called exceptions.

```python
try:
    x = int(input("Enter number: "))
except ValueError:
    print("That was no ordinary number!")
```

The code can react to multiple exceptions.

```python
try:
    1 / 0
except ZeroDivisionError:
    print("Division by zero is not allowed")
except Exception as e:
    print("A different error occurred:", e)
```

The try block can be supplemented with else and finally.

* the **else** block executes if no exception occurred in try
* **finally** always executes, useful for cleanup (file, database…)

```python
try:
    x = int(input("Enter number: "))
except ValueError:
    print("That's not a number!")
else:
    print("You entered the number:", x)
finally:
    print("This block is always executed.")
```

An exception can also be raised.

```python
raise IOError('Not enough space')
```

# Working with files

* We will work with files in later lessons.
* You should always handle file operations carefully and assume that some error may occur.

We will create a test file for the following examples.

```python
with open('file.txt', 'w') as f:
    f.write("Test content\n")
```

Bad practice. Error handling is completely missing.

```python
f = open ('file.txt', 'r')
contents = f.read()
f.close()
```

We can wrap the code with exception handling.

It's a bit better.

```python
f = open ('file.txt', 'r')
try:
    contents = f.read()
finally:
    f.close()
```

The best way to work with files is to use **with**.

This code always closes the file, even if an exception occurs

```python
with open ('file.txt', 'r') as f:
    contents = f.read()
```

# Custom exceptions

* We create a custom exception by inheriting from the `Exception` class (or its descendants).
* Useful when we want to more precisely name an error specific to our domain.

```python
class NegativeNumberError(Exception):
    """Exception for negative numbers."""
    pass

class RangeError(Exception):
    """Exception for a value outside the allowed range."""
    def __init__(self, value, minimum, maximum):
        super().__init__(f"Value {value} is not in range [{minimum}, {maximum}]")
        self.value = value

def square_root(x, minimum=0, maximum=1000):
    if x < 0:
        raise NegativeNumberError(f"Square root of a negative number is undefined: {x}")
    if x < minimum or x > maximum:
        raise RangeError(x, minimum, maximum)
    return x ** 0.5

try:
    print(square_root(9))       # 3.0
    print(square_root(100000))  # raises RangeError (100000 > 1000)
except RangeError as e:
    print("Range error:", e)

try:
    print(square_root(-4))      # raises NegativeNumberError
except NegativeNumberError as e:
    print("Error:", e)
```

# Re-raising an exception

`raise` without an argument inside an `except` block passes the caught exception further up. Useful when we want to log the exception or partially handle it, but let it continue propagating upward.

```python
def read_number(text):
    try:
        return int(text)
    except ValueError:
        print(f'[log] Could not convert "{text}" to a number.')
        raise            # the exception continues

try:
    read_number("abc")
except ValueError as e:
    print("Caught above:", e)
```
