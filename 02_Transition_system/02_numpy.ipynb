# The numpy library
The implementation of lists in Python is not efficient enough. NumPy is the fundamental library for scientific and numerical computing in Python, because it provides fast vectorized operations and efficient work with large data.

The numpy library is built for n-dimensional arrays, complex mathematical functions, random number generators, linear algebra routines, Fourier transforms, and more.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating an array
There are many ways to create a multidimensional array.
* From a list
* Generating a sequence of numbers
* A zero matrix, an identity matrix
* A random matrix, etc.

```python
a = np.array([1, 2, 3])       # array with elements 1, 2, 3
b = np.arange(9)              # array from 0 to 8
c = np.arange(1, 11, 2)       # array from 1 to 10 step 2 → [1, 3, 5, 7, 9]
d = np.zeros((4, 10))         # zero matrix 4×10, size passed as a tuple
e = np.ones((2, 3))           # unit matrix 2×3
f = np.random.rand(3, 5)      # random matrix; rand() takes individual args, not a tuple
g = np.eye(3, dtype=int)      # identity matrix
h = np.linspace(1, 11, 40)    # numbers evenly distributed within an interval
```

```python
print (a)
print (b)
print (c)
print (d)
print (e)
print (f)
print (g)
print (h)
```

## Data type
A matrix can contain data of various data types. The type of numbers is defined when it is created.

The chosen data type determines how much space the matrix takes up in memory, and also what range of numbers it can contain.

From a performance standpoint, it is advisable to choose the smallest suitable data type.

https://numpy.org/doc/stable/reference/arrays.dtypes.html

```python
np.dtype('i4')      # 32-bit signed integer
np.dtype('f8')      # 64-bit floating-point number
np.dtype('c16')     # 128-bit complex floating-point number
np.dtype('S25')     # 25-length zero-terminated bytes
np.dtype('U25')     # 25-character string
np.dtype('uint32')  # 32-bit unsigned integer
np.dtype('float64') # 64-bit floating-point number
np.dtype(float)     # Python-compatible floating-point number
np.dtype(int)       # Python-compatible integer
```

You can find out what data type a matrix contains using **dtype**.

```python
a = np.array([[1, 2], [3, 4]], dtype=float)
a.dtype
```

## Aggregate functions
Numpy contains functions that compute a single value from an array — sum, mean, minimum, maximum, standard deviation, etc.

For multidimensional arrays, the computation can be done over the whole matrix or along axes:
* `axis=0` — computes down the rows (the result has the shape of a single row)
* `axis=1` — computes across the columns (the result has the shape of a single column)

```python
print(f"sum:    {h.sum():.1f}")
print(f"mean:   {h.mean():.1f}")
print(f"std:    {h.std():.2f}")
print(f"min:    {h.min():.1f}  at index {h.argmin()}")
print(f"max:    {h.max():.1f}  at index {h.argmax()}")

# same functions work on 2D arrays; axis controls the direction
print("\nColumn means (axis=0):", f.mean(axis=0).round(2))
print("Row means   (axis=1):", f.mean(axis=1).round(2))
```

## Splitting a matrix
We will certainly find it useful later to have an operation that cuts some part out of a matrix. For example, we might want just certain columns or certain rows.

```python
print (f)
print (f.shape)
```

Let's start by taking the first row. Again, indexes are counted from 0.

```python
f[0]
```

If we give a further index, we slice by columns. We have already seen that negative numbers count lists from the end.

The following expression selects the last value from the first row.

```python
f[0][-1]
```

We can also slice columns and other submatrices from a matrix.
* f[:,2] selects the 3rd column, : means we want all the rows

```python
f[:,2]
```

* f[1:3,2:4] selects the 3rd and 4th column from the 2nd and 3rd row

```python
f[1:3,2:4]
```

Matrices can also be split using the split function, where instead of selecting specific submatrices, we make a cut through the matrix that divides it into two parts.

The following example separates the first column
* [1] - the dividing column
* axis=1 splits by columns

```python
print (f) 
```

```python
f1, f2 = np.split(f, [1], axis=1)    
print (f1)
print (" ")
print (f2)
```

Similarly, we can separate the first row.
* [1] the elements to the left will be in one matrix, the elements to the right (inclusive) in the other.
* axis=0 splits by rows

```python
f1, f2 = np.split(f, [1], axis=0)
print (f1)
print (" ")
print (f2)     
```

## Selection by condition (boolean indexing)
Instead of numeric indexes, elements can be selected by a condition. The comparison returns an array of True/False values, and numpy returns only the elements where it is True.

This way of selecting is used very often in practice — for example for filtering data or replacing outlier values.

```python
v = np.array([3, 7, 1, 9, 4, 6, 2, 8])
```

Select the elements greater than 5

```python
print(v[v > 5])
```

Conditions are combined using the characters & (and)  | (or)

```python
print(v[(v > 3) & (v < 8)])
```

Replace the values below the threshold with the number 0

```python
v2 = v.copy()
v2[v2 < 4] = 0
print(v2)
```

## Matrix operations
So far we have shown scalar operations with matrices. Numpy can of course also add and multiply matrices.

We can add matrices if they have the same dimensions.

```python
i=np.linspace (1, 10, 10)
j=np.linspace (11, 20, 10)
print (i)
print (j)
```

```python
print (i+j)
```

For matrix multiplication, the number of rows of one matrix must match the number of columns of the other matrix.

```python
A = np.arange (2, 14)
A = A.reshape ((3, 4))
B = np.arange (5, 25)
B = B.reshape ((4, 5))
print (A)
print (B)
```

```python
C=np.dot (A, B)
print (C)
```

## Transposition
Transposition swaps the rows and columns of a matrix. In numpy this is done using the `.T` attribute.

In linear algebra and machine learning, transposition is very common — for example when computing `X.T @ X`, which leads to the normal equations system of linear regression.

```python
print("A shape:", A.shape)
print(A)

print("\nA.T shape:", A.T.shape)
print(A.T)
```

## Creating a grid
Sometimes it is useful to create points in space arranged in a grid. We can then compute the value of some function for these points, in order to get a 3D graph.

The meshgrid function is used to create a grid; we pass it two vectors.

Creating a coordinate grid of a given size.

```python
xx=np.linspace(0, 1, 30)
yy=np.linspace(0, 1, 20)
print (xx)
print (yy)
```

```python
XX, YY = np.meshgrid (xx, yy)
print (XX.shape)
print (YY.shape)
```

```python
plt.scatter(XX, YY)
plt.show()
```

## Saving and loading an array to and from a file
If we process some multidimensional matrix, it can be useful to save it to a file for later processing.

We can save it to CSV. This way it will transfer more easily, for example to Excel.

And conversely, a matrix can be read from CSV.

```python
np.savetxt("foo.csv", xx, delimiter=",")
```

```python
arr = np.loadtxt("foo.csv", delimiter=",", dtype=str)
display(arr)
```

Text storage is good for transferring data between different programs. But the storage efficiency is poor.

That's why numpy allows saving matrices in its own optimized format.

```python
np.savez_compressed("arr", arr)
```

```python
arr2=np.load("arr.npz")
```
